```python
import os
```

```python
import numpy
distance_file = os.path.join('distance_data_headers.csv')
distances = numpy.genfromtxt(fname=distance_file, delimiter=',', dtype='unicode')
headers = distances [0,:]
data = distances [1:,:]
data = data.astype(numpy.float)
```

```python
print(headers)
```

```python
print(data)
```

```python
import matplotlib.pyplot as plt
```

```python
numcolumns = len(headers)
for num in range(1,numcolumns):
    sample = headers [num]
    plt.plot(data[:,num])

plt.xlabel('Simulation Frame')
plt.ylabel('Angstrom')
plt.legend(headers[num])

plt.savefig('two_samples.png', dpi=300)
```

```python
num_columns = len(headers)

print(headers)
print(F'The number of columns is {num_columns}')

for col in range(1, num_columns):
    print(F'The column number is {col}.')
    plt.figure()
    plt.plot(data[:,col], label=headers[col])
    
    plt.legend()
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance(Angstrom)')
    plt.savefig(F'{headers[col]}.png', dpi=200)
```
